## Import Loopring price data for 2021-2022 from CryptoCompare.com

Import 2 years of hourly Loopring (LRC) data.

There is a limit of 2000 hour calls per request so will have to make 14 separate requests.

```python
# API key for CryptoCompare.com

load_dotenv()
API_KEY = os.getenv('CRYPTOCOMPARE_API')
```

```python
# API connection & first batch of data
url = 'https://min-api.cryptocompare.com/data/v2/histohour?fsym=LRC&tsym=USD&limit=2000&api_key=API_KEY'
r = requests.get(url)
json_data = r.json()

# Save dataset for LRC in json format
with open("LRC_file.json", "w") as write_file:
    json.dump(json_data, write_file)
```

```python
# Daily price data for LRC
url = 'https://min-api.cryptocompare.com/data/v2/histoday?fsym=LRC&tsym=USD&limit=1000&api_key=YOUR-API-KEY'
r = requests.get(url)
json_data = r.json()

with open("LRC_daily.json", "w") as write_file:
    json.dump(json_data, write_file)

with open("LRC_daily.json", "r") as lrc_d:
    lrc_d = json.load(lrc_d)
    
lrc_daily_data = lrc_d['Data']['Data'][:]

# structure data into a dictionary
lrc_data = {}
lrc_time = []
lrc_open = []
lrc_close = []
lrc_volumeto = []

for i in lrc_daily_data:
    lrc_hr = datetime.datetime.fromtimestamp(int(i['time'])).strftime('%Y-%m-%d %H:%M')
    lrc_time.append(lrc_hr)
    lrc_open.append(i['open'])
    lrc_close.append(i['close'])
    lrc_volumeto.append(i['volumeto'])
    lrc_data={'date':lrc_time, 'open':lrc_open, 'close':lrc_close, 'volumeto':lrc_volumeto}

# to dataframe
lrc_df = pd.DataFrame(lrc_data)
print(lrc_df.head())

# save combined price data to csv
lrc_df.to_csv(r'lrc_daily.csv', index=False)
```
